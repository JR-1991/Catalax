```python
import pandas as pd

import catalax as ctx  # noqa
```

```python
# EnzymeML Dataset Import
enzml_ds = ctx.Dataset.from_enzymeml("./datasets/enzymeml.json")
enzml_ds.plot(measurement_ids=["m0", "m1"], figsize=(7, 3))
```

```python
# Croissant Dataset Import (see https://github.com/mlcommons/croissant)
croiss_ds = ctx.Dataset.from_croissant("./datasets/croissant_dataset.zip")
f = croiss_ds.plot(measurement_ids=[m.id for m in croiss_ds.measurements[:4]], figsize=(7, 3))
```

```python
# Import from DataFrame (using CSV here, but any DataFrame is valid)
meas = pd.read_csv("./datasets/measurements.csv")
inits = pd.read_csv("./datasets/inits.csv")

df_ds = ctx.Dataset.from_dataframe(name="From CSV", data=meas, inits=inits)
f = df_ds.plot(measurement_ids=[m.id for m in df_ds.measurements[:4]], figsize=(7, 3))
```
