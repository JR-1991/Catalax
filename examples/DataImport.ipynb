# Data Import Examples

This notebook demonstrates different ways to import data into Catalax datasets:

1. **Croissant Dataset Import**: Loading data from a Croissant format dataset (MLCommons standard)
2. **DataFrame Import**: Creating datasets from pandas DataFrames (CSV files in this example)

Both methods create Catalax Dataset objects that can be used for modeling and analysis.


```python
# Croissant Dataset Import (see https://github.com/mlcommons/croissant)
croiss_ds = ctx.Dataset.from_croissant("./datasets/croissant_dataset.zip")
f = croiss_ds.plot(measurement_ids=[m.id for m in croiss_ds.measurements[:4]])
```

```python
# Import from DataFrame (using CSV here, but any DataFrame is valid)
meas = pd.read_csv("./datasets/measurements.csv")
inits = pd.read_csv("./datasets/inits.csv")

df_ds = ctx.Dataset.from_dataframe(name="From CSV", data=meas, inits=inits)
f = df_ds.plot(measurement_ids=[m.id for m in df_ds.measurements[:4]])
```
